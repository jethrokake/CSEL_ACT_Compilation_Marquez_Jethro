1. import libraries

```python
import pandas as pd #data manipulation
import matplotlib.pyplot as plt
```

2. Loading dataset

```python
data = pd.read_csv('sales_data_2.csv')
data.head()
```

3. Data loading & initial Analysis

```python
statistics = data.describe
```

4. data manipulation

```python
product_revenue = data.groupby('Product')['Revenue'].sum()
high_revenue_product = product_revenue.idxmax()
average_quantity_per_day= data['Quantity Sold'].mean()
```

5.Data visualization

```python
product_revenue.plot(kind ='bar')
plt.xlabel('Product')
plt.ylaberl('Revenue')
plt.title('Revenue By Product')
plt.xticks(rotation = 45)
plt.show
```

6.analysist report

```python
report = """
Analysis report

1. Data Loading and Initial Analysis:
    Basic statitics of the dataset;
    {}
2. Data Manipulation:
    Total rvenue for each product:
    {}
    Product with the highest revenue: {}
    Average quantity sold per day:{:.2f}
3. Data Visualization:
    Bar char displaying revenue for each product is attached.
4. Insights:
    The product {} generates the highest revenue.
    On average, {:.2f} units are sold per day.

""". format(statistics,product_revenue,high_revenue_product,average_quantity_per_day,
            high_revenue_product,average_quantity_per_day)
print (report)
```

7.saving report to a file

```python
with open ('kake.txt','w') as f:
  f.write (report)
```
