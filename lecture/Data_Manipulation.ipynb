```python
import pandas as pd #Data manipulation
import numpy as np # numerical operation
import matplotlib.pyplot as plt #data visualization
from sklearn.linear_model import LinearRegression # DATA modeling
```

# 2. generating random dataset

```python
num_points = 50

X = np.random.randint(0,100, num_points)
Y = 2 * X + np.random.normal(0,10,num_points)
```

# 3. Creating Data framm and saving it as csv

```python
data = pd.DataFrame({'X': X, 'Y': Y})

data.to_csv('data.csv', index=False)

print(data.head())
```

# 4. Visualizing the data

```python
plt.scatter(data['X'], data['Y'])
plt.xlabel('independent Variable (X)')
plt.ylabel('dependent Variable (Y)')
plt.title ('scatter plot of data')
plt.show()
```

# 5. prepairing data for modeling

```python
X = data[['X']] #independent
Y = data[['Y']] #dependent
```

# 6. fitting thje linear reggresion model

```python
model = LinearRegression()

model.fit(X,Y)
```

7. Getting model coefficients
# New Section




```python
slope = model.coef_[0]

inetercept= model.intercept_

print('', slope)
print('Intercept', inetercept)
```

# 8.visualizing the linear re3ggresion line

```python
plt.scatter(X,Y)
plt.plot(X, model.predict(X), color = 'red')
plt.xlabel('dependent variable (X)')
plt.xlabel('independent variable (Y)')
plt.title('Correlation')
plt.show()
```

9. making prediction

```python
new_X = [[]]

predicted_Y = model.predict(new_X)

print('X= ', new_X)
print('predicted value nof Y: ', predicted_Y)
```
